```python
applications = pd.read_csv('../data/pruned_data.csv', low_memory=False)
applications
```

```python
applications = applications[['interest_rate','rate_spread','loan_term','intro_rate_period']]
applications
```

```python
for col in applications.columns[:-2]:
    applications = applications[(applications[col].notna()) & (applications[col] != 'Exempt')]
applications
```

```python
for col in applications.columns:
    applications[col] = applications[col].astype(float)
applications
```

```python
applications['prime_rate'] = round(applications['interest_rate'] - applications['rate_spread'],2)
applications
```
