```python
import pandas as pd
import numpy as np
```

```python
fixed_rates = pd.read_csv('../data/YieldTableFixed.txt',sep='|')
fixed_rates = fixed_rates.loc[260:312]
fixed_rates
```

```python
adj_rates = pd.read_csv('../data/YieldTableAdjustable.txt',sep='|')
adj_rates = adj_rates.loc[260:312]
adj_rates
```

```python
applications = pd.read_csv('../data/pruned_data.csv', low_memory=False, nrows = 200000)
applications
```

```python
applications = applications[['interest_rate','rate_spread','loan_term','intro_rate_period']]
applications
```

```python
for col in applications.columns[:-1]:
    applications = applications[(applications[col].notna()) & (applications[col] != 'Exempt')]
applications
```

```python
for col in applications.columns:
    applications[col] = applications[col].astype(float)
applications
```

```python
applications['prime_rate'] = round(applications['interest_rate'] - applications['rate_spread'],2)
applications
```

```python
applications['min_diff'] = np.nan
```

```python
def years_term(num):
    years, months = divmod(num,12)
    if months > 6:
        return int(years+1)
    elif years == 0:
        return int(years+1)
    else:
        return int(years)
```

```python
for ind, row in applications.iterrows():
    if np.isnan(row['intro_rate_period']):
        applications.loc[ind,'min_diff'] = min([abs(row['prime_rate']-rate) for rate in fixed_rates[str(years_term(row['loan_term']))]])
    else:
        applications.loc[ind,'min_diff'] = min([abs(row['prime_rate']-rate) for rate in adj_rates[str(years_term(row['intro_rate_period']))]])
```

```python
applications.sort_values('min_diff', ascending=False)
```
